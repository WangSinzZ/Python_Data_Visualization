```python
import pandas as pd
import plotly.express as px

filename = './date/all_month.json'

with open(filename, 'r', encoding='UTF-8') as file:
  all_eq_data = pd.read_json(file)
  all_eq_data1 = all_eq_data.drop(all_eq_data[all_eq_data.mag < 0].index, inplace=False) # 删除震级小于0的
  print(len(all_eq_data1))
  print(all_eq_data1)
```

```python
mags = all_eq_data1['mag']
# print(mags)
titles = all_eq_data1['place']
lats = all_eq_data1['latitude']
lons = all_eq_data1['longitude']


data = pd.DataFrame(
  data = zip (lons, lats, titles, mags), columns = ['经度', '纬度', '位置', '震级']
)

fig = px.scatter(
  data,
  x = '经度', 
  y = '纬度',
  range_x = [-200, 200],
  range_y = [-90, 90],
  width = 1440,
  height = 720,
  title = '全球地震散点图(2022-04-21 -- 2022-05-21)',
  size = '震级',
  color = '震级',
  size_max = 10.0,
)
fig.write_html('global_earthquakes.html')
fig.show()
```
